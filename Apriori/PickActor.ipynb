```python
from efficient_apriori import apriori
import csv
```

```python
director = u'周星驰'
file_nme = 'Apriori/'+director+'.csv'
lists = csv.reader(open(file_nme,'r',encoding='utf-8-sig'))
data = []
for names in lists:
    name_new = []
    for name in names:
        name_new.append(name.strip())
    data.append(name_new[1:])
itemsets,rules = apriori(data,min_support=0.1,min_confidence=1)
print(itemsets)
print('------'*20)
print(rules)
```
